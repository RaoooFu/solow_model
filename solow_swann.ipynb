```python
file_path_3='/Users/home/Desktop/Ironhack/Group Assignment/machine_learning/gross_capital_formation.xlsx'
df_investment=pd.read_excel(file_path_3)
df_investment
```

```python
df_investment = df_investment.rename(columns={'Country Name': 'Country'})
```

```python
df_investment
```

```python
# Convert all column names to strings first
df_investment.columns = df_investment.columns.map(str)

# Now select year columns
year_columns = [col for col in df_investment.columns if col.isnumeric()]

# Keep only 'Country' and year columns
df_investment_clean = df_investment[['Country'] + year_columns]

# Melt into long format
df_investment_long = df_investment_clean.melt(
    id_vars='Country',
    var_name='Year',
    value_name='Investment_rate'
)

# Convert year column to int
df_investment_long['Year'] = df_investment_long['Year'].astype(int)
df_investment_long
```

```python
def reshape_sheet(df, value_name):
    df = df.rename(columns={df.columns[0]: "Country"})  # Corrected here
    df_long = df.melt(id_vars='Country', var_name='Year', value_name=value_name)
    df_long['Year'] = df_long['Year'].astype(int)
    return df_long
```

```python
df_schooling
```

```python
gdp_df = reshape_sheet(dfs['world_real_gdp_growth'], 'GDPR')
gdp_pc_df = reshape_sheet(dfs['gdp_per_capita'], 'GDP_per_capita')
pop_df = reshape_sheet(dfs['population_growth'], 'Population_growth')

# Merge on Country and Year
merged_df = gdp_df.merge(gdp_pc_df, on=['Country', 'Year'], how='inner') \
                  .merge(pop_df, on=['Country', 'Year'], how='inner')

merged_df
```

```python
merged_df_schooling = merged_df.merge(df_schooling, on=['Country', 'Year'], how='left')
```

```python
merged_df_schooling
```

```python
merged_final=final_df = merged_df_schooling.merge(df_investment_long, on=['Country', 'Year'], how='left')
merged_final
```

```python
final_df.to_excel("merged_final.xlsx", index=False)
```
